```python
import pandas as pd
```

```python
q_metadata = pd.read_csv('data/metadata/question_metadata_task_3_4.csv')
print('QuestionId max :',q_metadata['QuestionId'].max())
print('QuestionId min :',q_metadata['QuestionId'].min())
print(q_metadata.info())
q_metadata.head(5)
```

```python
q_metadata['SubjectId'].value_counts()
```

```python
def extract_last_number(string):
    last_num_str = string.split(',')[-1].strip()
    return int(last_num_str[:-1])
for k in range(0,len(q_metadata)) :
    q_metadata.loc[k,'last_SubId']= extract_last_number(q_metadata.loc[k,'SubjectId'])

q_metadata.head()
```

```python
subject = pd.read_csv('data/metadata/subject_metadata.csv')
print(subject.info())
subject.head()
```

```python
question = pd.merge(q_metadata, subject, how='left', left_on='last_SubId', right_on='SubjectId')
question = question.drop(['SubjectId_y','Name','ParentId','SubjectId_x'],axis=1)
question.head()
```

```python
train = pd.read_csv('data/train_data/train_task_3_4.csv')
print('length :',len(train))
print('QuestionId max :',train['QuestionId'].max())
print('QuestionId min :',train['QuestionId'].min())
train.head()
```

```python
student = pd.read_csv('data/metadata/student_metadata_task_3_4.csv')
print('UserId max :',student['UserId'].max())
print('UserId min :',student['UserId'].min())
print(student.info())
student.head()
```

```python
UserAcc_data = {'UserId' : [],'UserAcc':[]}
UserAcc=pd.DataFrame(UserAcc_data)
# calc percent of correct for each question
for k,df in train.groupby('UserId'):
    temp={'UserId':k,'UserAcc':df['IsCorrect'].mean()}
    UserAcc=UserAcc.append(temp,ignore_index=True)
UserAcc.head()
```

```python
train_data= pd.merge(train, UserAcc, how='left', on='UserId')
train_data.head()
```

```python
Question_acc = {'QuestionId' : [],'AnsValmean':[],'RealAcc':[],'ExpAcc':[]}
question_acc=pd.DataFrame(Question_acc)
# calc percent of correct for each question
for k,df in train_data.groupby('QuestionId'):
    temp={'QuestionId':k,'AnsValmean':df['AnswerValue'].mean(),'RealAcc':df['IsCorrect'].mean(),'ExpAcc':df['UserAcc'].mean()}
    question_acc=question_acc.append(temp,ignore_index=True)
question_acc.head()
```

```python
x=pd.merge(question,question_acc,how='left',on='QuestionId')
x['AccGap']=question_acc['RealAcc']-question_acc['ExpAcc']
x.head()
```

```python
validation = pd.read_csv('data/test_data/quality_response_remapped_public.csv')
print('length :',len(validation))
validation.head()
```

```python
test = pd.read_csv('data/test_data/quality_response_remapped_private.csv')
print('length :',len(test))
test.head()
```

```python
total = validation.append(test)
total.head()
```

```python
final_data=pd.merge(x,total,how='right', left_on='QuestionId', right_on='Unnamed: 0')
final_data=final_data.set_index('QuestionId')
final_data=final_data.drop(['Unnamed: 0','Unnamed: 0.1'],axis=1)
final_data.head()
```
